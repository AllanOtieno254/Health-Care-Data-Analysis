# **Patients Age group by total visit**

```python
from re import A
# convert the admission date and date of birth to proper date format
# df['Admission Date']=df['Admission Date'].dtype
# print(df['Admission Date'])
# df['Date of Birth']=df['Date of Birth'].dtype
# print(df['Date of Birth'])
df["Admission Date"]=pd.to_datetime(df["Admission Date"])
df["Date of Birth"]=pd.to_datetime(df["Date of Birth"])

# calculating patients age as of the time of admission in year
df["Age"]=(df["Admission Date"]-df["Date of Birth"]).dt.days/365
df["Age"]=df["Age"].astype(int)
df["Age"].head()

# define age bins(grouping ages)
bins=[0,20,40,60,80,100]
labels=["0-20","21-40","41-60","61-80","81-100"]

# creating an age group using bins var defined
df["Age_group"]=pd.cut(df["Age"],bins=bins,labels=labels,right=False)
df['Age_group']

# count number of patients in each age group
# FIX: do not assign value_counts() to a column like this
# Instead, plot using Age_group directly
print(df['Age_group'])

import seaborn as sns
import matplotlib.pyplot as plt

fig,ax=plt.subplots(figsize=(20,8))
ax=sns.countplot(x="Age_group", data=df,label="Patients Age group by total visit")

for bars in ax.containers:
  ax.bar_label(bars)
```

# **Calculate the duration of each patient stay and create a group to show numbers of patients in each group**

# **What are the peak Days of the week on a monthly bases for admission and discharges?**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Convert to proper datetime
df['Admission Date'] = pd.to_datetime(df['Admission Date'])
df['Discharge Date'] = pd.to_datetime(df['Discharge Date'])

# Extract month and weekday names
df['Admission Month'] = df['Admission Date'].dt.strftime('%b')  # e.g., Jan, Feb
df['Admission Day'] = df['Admission Date'].dt.strftime('%a')    # e.g., Mon, Tue

# Define correct orders
month_order = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
               'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
day_order = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']

# Step 1: Create pivot table
admission_heatmap = df.pivot_table(
    index='Admission Day',
    columns='Admission Month',
    aggfunc='size',
    fill_value=0
)

# Step 2: Reorder rows and columns
admission_heatmap_sorted = admission_heatmap.reindex(index=day_order, columns=month_order)

# Step 3: Plot heatmap
plt.figure(figsize=(12, 6))
sns.heatmap(admission_heatmap_sorted, annot=True, fmt="d", cmap="viridis")
plt.title("Admissions by Day of Week and Month")
plt.xlabel("Month")
plt.ylabel("Day of Week")
plt.tight_layout()
plt.show()
```

# **What is the total number of patients admitted per wek,month and year? add filter to filter month by year**

```python
df['Admission Day']=df['Admission Date'].dt.strftime('%a')

#group data by the day name and count admission
admission_by_day=df.groupby("Admission Day").size()

day_order = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
admission_by_day=admission_by_day.reindex(day_order,fill_value=0)
admission_by_day
```

```python
#create a visual
admission_by_day.plot(kind="bar")
plt.title("Admissions by Day of Week")
plt.xlabel("Day of Week")
plt.ylabel("Number of Admissions")
plt.show()
```

```python
#bottom 3 days with high admission
top_5_admission_day=df['Admission Day'].value_counts().nsmallest(3)
top_5_admission_day
```

```python
# Filter using year of admission
selected_year = 2023
df_year = df[df['Admission Date'].dt.year == selected_year].copy()

# Adding month column
df_year['Admission Month'] = df_year['Admission Date'].dt.month

# Extracting month name
df_year['Month Name'] = df_year['Admission Date'].dt.month_name()

# Total admission on monthly basis
monthly_admission = (
    df_year
    .groupby(['Month Name', 'Admission Month'])
    .size()
    .reset_index(name='Total Patients')
    .sort_values('Admission Month')
)
print(f"Total patients admitted in the year {selected_year}")
monthly_admission
```

```python
#create line chart
import seaborn as sns
import matplotlib.pyplot as plt

# Set Seaborn style
sns.set(style='whitegrid')

# Create line chart
plt.figure(figsize=(10, 6))
sns.lineplot(
    data=monthly_admission.sort_values('Admission Month'),
    x='Month Name',
    y='Total Patients',
    marker='o'
)

# Improve readability
plt.title(f'Monthly Patient Admissions in {selected_year}')
plt.xlabel('Month')
plt.ylabel('Total Patients')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
