# **Patients Age group by total visit**

```python
from re import A
# convert the admission date and date of birth to proper date format
# df['Admission Date']=df['Admission Date'].dtype
# print(df['Admission Date'])
# df['Date of Birth']=df['Date of Birth'].dtype
# print(df['Date of Birth'])
df["Admission Date"]=pd.to_datetime(df["Admission Date"])
df["Date of Birth"]=pd.to_datetime(df["Date of Birth"])

# calculating patients age as of the time of admission in year
df["Age"]=(df["Admission Date"]-df["Date of Birth"]).dt.days/365
df["Age"]=df["Age"].astype(int)
df["Age"].head()

# define age bins(grouping ages)
bins=[0,20,40,60,80,100]
labels=["0-20","21-40","41-60","61-80","81-100"]

# creating an age group using bins var defined
df["Age_group"]=pd.cut(df["Age"],bins=bins,labels=labels,right=False)
df['Age_group']

# count number of patients in each age group
# FIX: do not assign value_counts() to a column like this
# Instead, plot using Age_group directly
print(df['Age_group'])

import seaborn as sns
import matplotlib.pyplot as plt

fig,ax=plt.subplots(figsize=(20,8))
ax=sns.countplot(x="Age_group", data=df,label="Patients Age group by total visit")

for bars in ax.containers:
  ax.bar_label(bars)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Combine into a DataFrame
top3_df = roundpct.reset_index()
top3_df.columns = ['Disease', 'Percentage']

# Plot
plt.figure(figsize=(8,5))
ax = sns.barplot(x="Disease", y="Percentage", data=top3_df, palette="Set2")
plt.title("Top 3 Diseases as Percentage")
plt.ylabel("Percentage (%)")
plt.xlabel("Disease")
plt.ylim(0, 100)

# Add bar labels using containers
for bars in ax.containers:
    ax.bar_label(bars, fmt='%d%%')

plt.show()
```

# **Are certain diseases more prevalent in one gender compared to other**

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Crosstab for disease vs gender
gender_disease = pd.crosstab(df["Disease"], df["Gender"])

# Reset index to convert to long-form DataFrame for plotting
gender_disease_plot = gender_disease.reset_index().melt(id_vars='Disease', var_name='Gender', value_name='Count')

# Plot
plt.figure(figsize=(12, 6))
ax = sns.barplot(x="Disease", y="Count", hue="Gender", data=gender_disease_plot, palette="Set2")

plt.title("Disease Distribution by Gender", fontsize=14)
plt.xlabel("Disease")
plt.ylabel("Number of Patients")
plt.xticks(rotation=45, ha='right')
plt.legend(title="Gender")
plt.tight_layout()

# Optional: add labels
for container in ax.containers:
    ax.bar_label(container, fmt='%d', label_type='edge')

plt.show()
```
